```python
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
file = './hackathon/Book2.csv'
data = pd.read_csv(file, header=0)
data = data.drop(columns="num")
data.head()
```

```python
#turn hex to rgb
dt = data.copy()
def hextorgb(st):
    h=st.lstrip('#')
    return tuple(int(h[i:i+2], 16) for i in (0, 2, 4))
dt['rgb']=data['hex'].map(lambda x: hextorgb(x))
#dt = dt.drop(columns="hex")
#justrgb = dt.drop(columns="l")
print(dt[:3])
print(dt.iat[1,1])
print(type(dt.iat[1,3]))
print(dt.shape)
print(type(dt))
#datanp = justrgb.to_numpy()
#print(datanp.shape)
```

```python
#turn rgb to r,g,b(between 0 to 1)
rgbl = dt.copy()
rgbl['r']=rgbl['rgb'].map(lambda x: x[0]/255)
rgbl['g']=rgbl['rgb'].map(lambda x: x[1]/255)
rgbl['b']=rgbl['rgb'].map(lambda x: x[2]/255)
rgbl.head()
```

```python
#turn rgb to hsv
import colorsys
def tohsv(c):
    return(colorsys.rgb_to_hsv(c[0]/255,c[1]/255,c[2]/255))
data2 = rgbl.copy()
data2['hsv'] = rgbl['rgb'].map(lambda x: tohsv(x))
data2.head()
```

```python
#turn hsv to h s v
data2['hue']=data2['hsv'].map(lambda x: x[0])
data2['sat']=data2['hsv'].map(lambda x: x[1])
data2['val']=data2['hsv'].map(lambda x: x[2])
data2.head()
```
